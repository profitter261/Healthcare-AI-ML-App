```python
# ---------------- FAQ Knowledge Base ----------------
faq_data = {
    "What is Healthcare AI?": 
        "Healthcare AI refers to the use of artificial intelligence technologies such as machine learning and natural language processing to assist in diagnosis, treatment planning, drug development, and patient monitoring.",
    
    "How can AI help doctors?": 
        "AI can analyze large amounts of medical data, detect patterns in symptoms or scans, suggest possible diagnoses, recommend treatments, and automate administrative tasks to save time for doctors.",
    
    "Is my medical data safe with AI?": 
        "Healthcare AI systems are required to follow strict data privacy and security regulations such as HIPAA. Always use trusted providers who ensure encryption and compliance.",
    
    "Can AI replace doctors?": 
        "No. AI assists doctors but cannot replace human judgment, empathy, and decision-making in healthcare. AI is a supportive tool, not a replacement.",
    
    "How does AI diagnose diseases?": 
        "AI models are trained on large datasets of medical records, images, and lab results. When given new patient data, AI compares it with learned patterns to suggest possible conditions.",
    
    "What are the benefits of Healthcare AI?": 
        "Faster diagnosis, personalized treatment, reduced costs, early detection of diseases, and improved patient outcomes are the major benefits of Healthcare AI."
}
```

```python
# ---------------- Helper Function ----------------
def faq_chatbot():
    print("\n🤖 Welcome to the Healthcare AI FAQ Chatbot!")
    
    while True:
        print("\n📌 Please select a question to know more (or enter 0 to exit):")
        for i, question in enumerate(faq_data.keys(), start=1):
            print(f"{i}. {question}")
        
        choice = input("\nYour choice: ")
        
        if choice.isdigit():
            choice = int(choice)
            if choice == 0:
                print("\n👋 Thank you for using the Healthcare AI FAQ Chatbot. Stay healthy!")
                break
            elif 1 <= choice <= len(faq_data):
                question = list(faq_data.keys())[choice - 1]
                print(f"\n❓ {question}")
                print(f"💡 {faq_data[question]}")
            else:
                print("\n⚠️ Invalid choice. Please try again.")
        else:
            print("\n⚠️ Please enter a valid number.")
```

```python
# ---------------- Run Chatbot ----------------
faq_chatbot()
```
